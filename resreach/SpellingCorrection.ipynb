```python
import re
import numpy as np
from unidecode import unidecode

# Định nghĩa hàm remove_random_accent
def remove_random_accent(text, ratio=0.15):
    words = text.split()
    mask = np.random.random(size=len(words)) < ratio
    for i in range(len(words)):
        if mask[i]:
            words[i] = unidecode(words[i])
    return ' '.join(words)

# Định nghĩa hàm replace_accent_chars
chars_regrex = '[aàảãáạăằẳẵắặâầẩẫấậAÀẢÃÁẠĂẰẲẴẮẶÂẦẨẪẤẬoòỏõóọôồổỗốộơờởỡớợOÒỎÕÓỌÔỒỔỖỐỘƠỜỞỠỚỢeèẻẽéẹêềểễếệEÈẺẼÉẸÊỀỂỄẾỆuùủũúụưừửữứựUÙỦŨÚỤƯỪỬỮỨỰiìỉĩíịIÌỈĨÍỊyỳỷỹýỵYỲỶỸÝỴnNvVmMCG]'
same_chars = {
    'a': ['á', 'à', 'ả', 'ã', 'ạ', 'ấ', 'ầ', 'ẩ', 'ẫ', 'ậ', 'ắ', 'ằ', 'ẳ', 'ẵ', 'ặ'],
    'A': ['Á','À','Ả','Ã','Ạ','Ấ','Ầ','Ẩ','Ẫ','Ậ','Ắ','Ằ','Ẳ','Ẵ','Ặ'],
    'O': ['Ó','Ò','Ỏ','Õ','Ọ','Ô','Ố','Ồ','Ổ','Ỗ','Ộ','Ơ','Ớ','Ờ','Ở','Ỡ','Ợ','Q'],
    'o': ['ó', 'ò', 'ỏ', 'õ', 'ọ', 'ô', 'ố', 'ồ', 'ổ', 'ỗ', 'ộ', 'ơ','ớ', 'ờ', 'ở', 'ỡ', 'ợ', 'q'],
    'e': ['é', 'è', 'ẻ', 'ẽ', 'ẹ', 'ế', 'ề', 'ể', 'ễ', 'ệ', 'ê'],
    'E': ['É', 'È', 'Ẻ', 'Ẽ', 'Ẹ', 'Ế', 'Ề', 'Ể', 'Ễ', 'Ệ', 'Ê'],
    'u': ['ú', 'ù', 'ủ', 'ũ', 'ụ', 'ứ', 'ừ', 'ử', 'ữ', 'ự', 'ư'],
    'U': ['Ú', 'Ù', 'Ủ', 'Ũ', 'Ụ', 'Ứ', 'Ừ', 'Ử', 'Ữ', 'Ự', 'Ư'],
    'i': ['í', 'ì', 'ỉ', 'ĩ', 'ị'],
    'I': ['Í', 'Ì', 'Ỉ', 'Ĩ', 'Ị'],
    'y': ['ý', 'ỳ', 'ỷ', 'ỹ', 'ỵ', 'v'],
    'Y': ['Ý', 'Ỳ', 'Ỷ', 'Ỹ', 'Ỵ', 'V'],
    'n': ['m'],
    'N': ['N'],
    'v': ['y'],
    'V': ['Y'],
    'm': ['n'],
    'M': ['N'],
    'C': ['G'],
    'G': ['C']
}

def _char_regrex(text):
    match_chars = re.findall(chars_regrex, text)
    return match_chars

def _random_replace(text, match_chars):
    replace_char = match_chars[np.random.randint(low=0, high=len(match_chars), size=1)[0]]
    insert_chars = same_chars[unidecode(replace_char)]
    insert_char = insert_chars[np.random.randint(low=0, high=len(insert_chars), size=1)[0]]
    text = text.replace(replace_char, insert_char, 1)
    return text

def change(text):
    match_chars = _char_regrex(text)
    if len(match_chars) == 0:
        return text
    text = _random_replace(text, match_chars)
    return text

def replace_accent_chars(text, ratio=0.15):
    words = text.split()
    mask = np.random.random(size=len(words)) < ratio
    for i in range(len(words)):
        if mask[i]:
            words[i] = change(words[i])
    return ' '.join(words)

# Định nghĩa hàm remove_random_space
def remove_random_space(text, ratio=0.1):
    words = text.split()
    n_words = len(words)
    new_words = []
    i = 0
    while i < n_words:
        if i < n_words - 1 and np.random.random() < ratio:
            new_words.append(words[i] + words[i + 1])
            i += 2
        else:
            new_words.append(words[i])
            i += 1
    return ' '.join(new_words)

# Định nghĩa hàm chuyển đổi teencode
teencode_dict = {
    "không": "ko", "anh": "a", "em": "e", "biết": "bít", "giờ": "h", "gì": "j", "muốn": "mún", "học": "hok", "yêu": "iu",
    "chồng": "ck", "vợ": "vk", "ông": "ô", "được": "đc", "tôi": "t", "thích": "thik", "vì": "vì", "như": "z", "không": "k",
    "bạn": "b", "mình": "m", "gặp": "găp", "đây": "đây", "mới": "ms"
}

def replace_teencode(text, ratio=0.1):
    words = text.split()
    mask = np.random.random(size=len(words)) < ratio
    for i in range(len(words)):
        if mask[i] and words[i] in teencode_dict:
            words[i] = teencode_dict[words[i]]
    return ' '.join(words)

# Định nghĩa hàm add_noise
def add_noise(text):
    text = remove_random_accent(text, 0.15)
    text = replace_accent_chars(text, 0.15)
    text = remove_random_space(text, 0.1)
    text = replace_teencode(text, 0.1)
    return text
```
